Hacer un data cleaning total, chequear qué hacer con los valores faltantes de cada columna y limpiar el data frame hasta que no queden celdas vacías

```python
df["age"] = df["age"].fillna(df["age"].mean())
df["age"].isnull().sum()
```

```python
df["fare"] = df["fare"].fillna(method="bfill")
df["fare"].isnull().sum()
```

```python
df["cabin"] = df["cabin"].fillna("Desconocido")
df["cabin"].isnull().sum()
```

```python
df["embarked"].value_counts()
```

```python
df["embarked"] = df["embarked"].fillna("S")
df["embarked"].isnull().sum()
```

```python
cant = len(df.index) # df.index = rango de índices
cant_nulos_body = df["body"].isnull().sum()
rep = (cant_nulos_body/cant) * 100 
rep
```

```python
df.drop(["body"], axis=1, inplace=True)
```

```python
df["home.dest"].value_counts()
```

```python
df["home.dest"] = df["home.dest"].fillna("Desconocido")
df["home.dest"].isnull().sum()
```

```python
df["boat"].value_counts()
df["boat"] = df["boat"].fillna("13")
df["boat"].isnull().sum()
```

```python
df.isnull().sum()
```

Averiguar cuál fue la media de las edades e imprimir “El promedio de edades fue: <media>”

```python
promedio_edad = df['age'].mean()

print(f'El promedio de edades fue {round(promedio_edad, 2)}')
```

 Crear una tabla con la cantidad de sobrevivientes divididos por sexo

Crear una tabla con la cantidad de sobrevivientes mayores de 20

```python
df_filtrado = df.query('age>20')

df_filtrado
```

```python
df_edades = df[['age']]

mayores_df = df_edades.query('age>20')

mayores_df
```

Crear una tabla con la cantidad de sobrevivientes mujeres mayores de 20

```python
df_mujeres_mayores = df_filtrado.pivot_table(index='sex', aggfunc={'survived':sum})
df_mujeres_mayores
```
